```python
import matplotlib as mpl
import matplotlib.font_manager as fm
from matplotlib import font_manager as fm, pyplot as plt

!wget https://github.com/trishume/OpenTuringCompiler/blob/master/stdlib-sfml/fonts/Times%20New%20Roman.ttf
!wget https://github.com/matomo-org/travis-scripts/blob/master/fonts/Arial.ttf 

font_files = fm.findSystemFonts()

# Go through and add each to Matplotlib's font cache.
for font_file in font_files:
    fm.fontManager.addfont(font_file)

fm.fontManager.ttflist += fm.createFontList(['Times New Roman.ttf'])

# Use your new font on all your plots.
plt.rc('font', family='serif')
```

```python
from tensorflow.keras.models import load_model
import tensorflow as tf
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import np_utils
from keras.datasets import cifar10
from keras.optimizers import RMSprop
from keras.layers import Conv2D, MaxPooling2D, Flatten, AveragePooling2D

from google.colab import drive
drive.mount('/content/drive')

# the data, shuffled and split between train and val sets
# Here we are using the official test set as our validation set, in further
# tutorials, test and validation splits will be explained properly.
# Hence, even though the variables are `x_test` and `y_test`, they represent our validation set
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

print('Image shape: {0}'.format(x_train.shape[1:]))
print('Total number of training samples: {0}'.format(x_train.shape[0]))
print('Total number of validation samples: {0}'.format(x_test.shape[0]))

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train /= 255
x_test /= 255

y_train_class = np_utils.to_categorical(y_train, 10)
y_test_class = np_utils.to_categorical(y_test, 10)

from keras.models import load_model

model_layers_4 = load_model('/content/drive/MyDrive/dl-cw/layer_4')
model_layers_5 = load_model('/content/drive/MyDrive/dl-cw/layer_5')
model_layers_6 = load_model('/content/drive/MyDrive/dl-cw/layer_6')

model_filter_1 = load_model('/content/drive/MyDrive/dl-cw/layer_3')
model_filter_2 = load_model('/content/drive/MyDrive/dl-cw/filter_2')
model_filter_3 = load_model('/content/drive/MyDrive/dl-cw/layer_6')
model_filter_4 = load_model('/content/drive/MyDrive/dl-cw/filter_4')
model_filter_5 = load_model('/content/drive/MyDrive/dl-cw/filter_5')

model_dense_1 = load_model('/content/drive/MyDrive/dl-cw/filter_4')
model_dense_2 = load_model('/content/drive/MyDrive/dl-cw/dense_2')
model_dense_3 = load_model('/content/drive/MyDrive/dl-cw/dense_3')
model_dense_4 = load_model('/content/drive/MyDrive/dl-cw/dense_4')
model_dense_5 = load_model('/content/drive/MyDrive/dl-cw/dense_5')
model_dense_6 = load_model('/content/drive/MyDrive/dl-cw/dense_6')

model_pool_1 = load_model('/content/drive/MyDrive/dl-cw/dense_5')
model_pool_2 = load_model('/content/drive/MyDrive/dl-cw/pool_2')
model_pool_3 = load_model('/content/drive/MyDrive/dl-cw/pool_3')
model_pool_4 = load_model('/content/drive/MyDrive/dl-cw/pool_4')

model_act_1 = load_model('/content/drive/MyDrive/dl-cw/dense_5')
model_act_2 = load_model('/content/drive/MyDrive/dl-cw/act_2')
model_act_3 = load_model('/content/drive/MyDrive/dl-cw/act_3')
model_act_4 = load_model('/content/drive/MyDrive/dl-cw/act_4')

score4 = model_layers_4.evaluate(x_test, y_test_class, verbose=0)
score5 = model_layers_5.evaluate(x_test, y_test_class, verbose=0)
score6 = model_layers_6.evaluate(x_test, y_test_class, verbose=0)

score7 = model_filter_1.evaluate(x_test, y_test_class, verbose=0)
score8 = model_filter_2.evaluate(x_test, y_test_class, verbose=0)
score9 = model_filter_3.evaluate(x_test, y_test_class, verbose=0)
score10 = model_filter_4.evaluate(x_test, y_test_class, verbose=0)
score11 = model_filter_5.evaluate(x_test, y_test_class, verbose=0)

score12 = model_dense_1.evaluate(x_test, y_test_class, verbose=0)
score13 = model_dense_2.evaluate(x_test, y_test_class, verbose=0)
score14 = model_dense_3.evaluate(x_test, y_test_class, verbose=0)
score15 = model_dense_4.evaluate(x_test, y_test_class, verbose=0)
score16 = model_dense_5.evaluate(x_test, y_test_class, verbose=0)
score17 = model_dense_6.evaluate(x_test, y_test_class, verbose=0)

score18 = model_pool_1.evaluate(x_test, y_test_class, verbose=0)
score19 = model_pool_2.evaluate(x_test, y_test_class, verbose=0)
score20 = model_pool_3.evaluate(x_test, y_test_class, verbose=0)
score21 = model_pool_4.evaluate(x_test, y_test_class, verbose=0)

score22 = model_act_1.evaluate(x_test, y_test_class, verbose=0)
score23 = model_act_2.evaluate(x_test, y_test_class, verbose=0)
score24 = model_act_3.evaluate(x_test, y_test_class, verbose=0)
score25 = model_act_4.evaluate(x_test, y_test_class, verbose=0)
```

```python
layers = ["1 layer", "2 layers", "3 layers", "16 32\n64", "32 64\n128", "64 128\n256", "128 256\n512", "256 512\n1024", "50", "128", "256", "512", "50 50", "100 50", "Max\nw/ Avg", "Max", "Avg", "No pool\n& padding", "ReLU", "Sigmoid", "ELU", "SELU"]
val_loss = [score4[0], score5[0], score6[0], score7[0], score8[0], score9[0], score10[0], score11[0], score12[0], score13[0], score14[0], score15[0], score16[0], score17[0], score18[0], score19[0], score20[0], score21[0], score22[0], score23[0], score24[0], score25[0]]
val_acc = [score4[1], score5[1], score6[1], score7[1], score8[1], score9[1], score10[1], score11[1], score12[1], score13[1], score14[1], score15[1], score16[1], score17[1], score18[1], score19[1], score20[1], score21[1], score22[1], score23[1], score24[1], score25[1]]

print(len(layers))
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, round(y[i],3), round(y[i],3), ha = 'center', fontsize=12)

bar1 = plt.figure(figsize=(22,8))
barplot = plt.bar(layers[0:3], val_acc[0:3], color='maroon', label="No. of Conv. Layers")
barplot = plt.bar(layers[3:8], val_acc[3:8], color='green', label="No. of Filters")
barplot = plt.bar(layers[8:14], val_acc[8:14], color='blue', label="FC Layer(s)")
barplot = plt.bar(layers[14:18], val_acc[14:18], color='red', label="Pooling")
barplot = plt.bar(layers[18:23], val_acc[18:23], color='magenta', label="Activation")
plt.title("Validation accuracy", fontsize = 14)
addlabels(layers, val_acc)
plt.xticks(wrap=True, fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylim([0.4, 0.85])
plt.legend(loc=2, fontsize=12)
plt.show()

bar2 = plt.figure(figsize=(22,8))
barplot2 = plt.bar(layers[0:3], val_loss[0:3], color='maroon', label="No. of Conv. Layers")
barplot2 = plt.bar(layers[3:8], val_loss[3:8], color='green', label="No. of Filters")
barplot2 = plt.bar(layers[8:14], val_loss[8:14], color='blue', label="FC Layer(s)")
barplot2 = plt.bar(layers[14:18], val_loss[14:18], color='red', label="Pooling")
barplot2 = plt.bar(layers[18:23], val_loss[18:23], color='magenta', label="Activation")
plt.title("Validation loss", fontsize = 14)
addlabels(layers, val_loss)
plt.xticks(wrap=True, fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylim([0.5, 2.5])
plt.legend(loc=2,fontsize = 12)
plt.show()

bar1.savefig('/content/drive/MyDrive/dl-cw/barplot1.png')
bar2.savefig('/content/drive/MyDrive/dl-cw/barplot2.png')
```
